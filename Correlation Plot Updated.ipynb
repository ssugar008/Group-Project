```python
import pandas as pd
final = pd.read_csv('final_dataframe_with_vote_count_and_likes_2015_2016.csv', encoding ='ISO-8859-1')
```

```python
final.columns = ['useless', 'Actor1 Id', 'Actor1 Name', 'Actor2 Id', 'Actor2 Name',
       'Actor3 Id', 'Actor3 Name', 'Popularity', 'Release Date', 'Text',
       'Title', 'useless1', 'useless2', 'adult', 'backdrop_path',
       'belongs_to_collection', 'budget', 'genres', 'homepage', 'id',
       'imdb_id', 'original_language', 'original_title', 'overview',
       'poster_path', 'production_companies', 'production_countries',
       'release_date', 'revenue', 'runtime', 'spoken_languages', 'status',
       'tagline', 'video', 'vote_average', 'vote_count', 'viewCount',
       'dislikeCount', 'likeCount']
final = final.drop('useless',1)
final = final.drop('useless1',1)
final = final.drop('useless2',1)
```

```python
final['Length of all Comments'] = len(final['Text'])
for i in range(1028):
    final['Length of all Comments'].iloc[i] = len(final['Text'].iloc[i])
final
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr = final.corr()
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values, annot=True)
plt.show()
```
